## Add, multiply, divide and square root of arange

````python
vec5=np.arange(3,8)
vec6=np.arange(1,6)
print(vec5)
print('#-------------------------------break------------------------------------------------#')
print(vec6)
print('#-------------------------------break------------------------------------------------#')
print(vec5+vec6)
print('#-------------------------------break------------------------------------------------#')
print(vec5*vec6)
print('#-------------------------------break------------------------------------------------#')
print(1/vec5)
print('#-------------------------------break------------------------------------------------#')
print(np.sqrt(vec5*vec6))
print('#-------------------------------break------------------------------------------------#')
````

## Random Vectors

````python
rand_vec1=np.random.randn(19)
print(rand_vec1)
print('#-------------------------------break------------------------------------------------#')
print(rand_vec1[6])
print('#-------------------------------break------------------------------------------------#')
print(rand_vec1[4:9])
````

## Non Consecutive entries in np.arange

````python
print(np.arange(0,15,3))
print(rand_vec1[np.arange(0,15,3)])
````

## Changing some values in an array(vector)

````python
print(rand_vec1)
print('#-------------------------------break------------------------------------------------#')
rand_vec1[3:5]=4
print('#-------------------------------break------------------------------------------------#')
print(rand_vec1)
print('#-------------------------------break------------------------------------------------#')
rand_vec1[3:5]=[1,2]
print('#-------------------------------break------------------------------------------------#')
print(rand_vec1)
````

````python
print(rand_mat1)
print('#-------------------------------break------------------------------------------------#')
rand_mat1[1:3,3:5]=0
print('#-------------------------------break------------------------------------------------#')
print(rand_mat1)
````

## Sub-Matrices

````python
sub_mat1=rand_mat1[0:2,0:3]
print(sub_mat1)
print('#-------------------------------break------------------------------------------------#')
sub_mat1[:]=3
print(sub_mat1)
print('#-------------------------------break------------------------------------------------#')
print(rand_mat1)
````

````python
sub_mat2=rand_mat1[0:2,0:3].copy()
sub_mat2[:]=99
print(sub_mat2)
print('#-------------------------------break------------------------------------------------#')
print(rand_mat1)
````

## Indexing using Logicals

````python
rand_vec8=np.random.randn(15)
print('#-------------------------------break------------------------------------------------#')
print(rand_vec8)
print('#-------------------------------break------------------------------------------------#')
print(rand_vec8>0)
print('#-------------------------------break------------------------------------------------#')
print(rand_vec8[rand_vec8>0])
print('#-------------------------------break------------------------------------------------#')
````

````python
print(random_mat)
print('#-------------------------------break------------------------------------------------#')
print(random_mat>0)
print('#-------------------------------break------------------------------------------------#')
print(random_mat[random_mat>0])
````
